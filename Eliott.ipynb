```python
from plotnine import (
    ggplot, geom_point, aes, stat_smooth,
    facet_wrap, scale_color_manual
)
from plotnine.data import mtcars
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as mcolors

# Get 3 colors from the viridis colormap
viridis_cmap = plt.get_cmap('viridis')
viridis_colors = [mcolors.to_hex(viridis_cmap(i / 2)) for i in range(3)]

(
    ggplot(mtcars, aes("wt", "mpg", color="factor(gear)"))
    + geom_point()
    + stat_smooth(method="lm")
    + facet_wrap("~gear")
    + scale_color_manual(values=viridis_colors)
)
```
